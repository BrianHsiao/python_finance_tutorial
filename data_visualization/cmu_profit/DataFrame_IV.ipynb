```python
GOOG=pd.read_csv('GOOG.csv')
GOOG.set_index(pd.to_datetime(GOOG['Date'],format='%Y-%m-%d'),inplace=True)
GOOG=GOOG[['Adj Close','Volume']]
GOOG.rename(columns={'Adj Close':'GOOG_P','Volume':'GOOG_V'},inplace=True)
GOOG.head()
```

```python
MSFT=pd.read_csv('MSFT.csv')
MSFT.set_index(pd.to_datetime(MSFT['Date'],format='%Y-%m-%d'),inplace=True)
MSFT=MSFT[['Adj Close','Volume']]
MSFT.rename(columns={'Adj Close':'MSFT_P','Volume':'MSFT_V'},inplace=True)
MSFT.head()
```

```python
df=pd.merge(GOOG,MSFT,left_index=True,right_index=True,how='inner')
df.head()
```

```python
df[['GOOG_P','MSFT_P']].plot()
```

```python
df_price=df[['GOOG_P','MSFT_P']]
df_dr=np.log(df_price).diff(1)
df_dr.head()
```

```python
df_dr.dropna(inplace=True)
df_TR=np.exp(np.cumsum(df_dr))-1
df_TR.plot()
plt.show()
```
