```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
class_1=['Finance','Industry','IT']
class_2=['Banks','Securities','Cars','Oils','Computers','5G']

val_1=[80,70,100]
val_2=[30,50,35,35,30,70]
```

```python
fig,ax=plt.subplots(figsize=(5,5))

colors=['lightblue','skyblue','lightcoral','orange','lightgreen','limegreen']

pie1,c1_text,n1_text=ax.pie(val_1,labels=class_1,autopct='%.2f%%',radius=0.7,labeldistance=0.2,pctdistance=0.7,wedgeprops=dict(width=0.3,edgecolor='w'))
ax.pie(val_2,labels=class_2,autopct='%.2f%%',radius=1,pctdistance=0.9,wedgeprops=dict(width=0.3,edgecolor='w'),colors=colors)

for t in c1_text:
    t.set_size(13)

for n in n1_text:
    n.set_size(13)

ax.set(title='class', aspect='equal')

plt.show()
```
