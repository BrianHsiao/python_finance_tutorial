```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
products=['A','B','C','D']
revenue=[20,50,40,70]
revenue_2019=[30,100,20,70]
colors=['yellow','pink','blue','grey']
explode=[0,0,0,0.2]
```

```python
fig,axis=plt.subplots(1,2)
axis[0].pie(revenue,labels=products,autopct='%.0f%%',radius=1,colors=colors,explode=None,shadow=True)
axis[0].set_title('2018')
axis[1].pie(revenue_2019,labels=products,autopct='%.0f%%',radius=1,colors=colors,explode=None,shadow=True)
axis[1].set_title('2019')
plt.show()
```
