```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
GOOG=pd.read_csv('GOOG.csv')
GOOG.set_index(pd.to_datetime(GOOG['Date'],format='%Y-%m-%d'),inplace=True)
GOOG.head()
```

```python
MSFT=pd.read_csv('MSFT.csv')
MSFT.set_index(pd.to_datetime(MSFT['Date'],format='%Y-%m-%d'),inplace=True)
MSFT.head()
```

```python
fig,ax=plt.subplots(figsize=(10,5))
ax.plot(GOOG['Adj Close'],color='skyblue',label='GOOG')
ax.set_ylabel('GOOG',color='skyblue',fontsize=20)
ax.tick_params(axis='y',labelcolor='skyblue')
plt.grid()
ax.legend(loc='upper left')

ax2=ax.twinx()
ax2.plot(MSFT['Adj Close'],color='pink',label='MSFT')
ax2.set_ylabel('MSFT',color='pink',fontsize=20)
ax2.tick_params(axis='y',labelcolor='pink')
ax2.legend(loc='upper right')

plt.title('GOOG v.s. MSFT', color='limegreen')



plt.show()
```
