```python
import pandas as pd
import numpy as np
from numpy.random import normal, uniform
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
return1=normal(size=100)*10
risk1=uniform(0,0.5,size=100)

return2=normal(0.5,size=100)*10
risk2=uniform(0.3,1,size=100)

return3=normal(-2,size=100)*10
risk3=uniform(0,1,size=100)
```

```python
plt.scatter(risk1,return1,label='Stock 1')
plt.scatter(risk2,return2,label='Stock 2')
plt.scatter(risk3,return3,label='Stock 3')

plt.xlabel('risk')
plt.ylabel('return')
plt.legend(loc='lower right')

plt.show()
```

```python
df=pd.read_csv('risk_return.csv')
df
```

```python
fig, ax=plt.subplots(figsize=(10,5))

ax.scatter(df['risk'],df['return'])
ax.scatter(df['risk'][3],df['return'][3],color='red',marker='*',s=500,label='Gold ETF')

plt.legend()
plt.xlabel('risk',size=15)
plt.ylabel('return',size=15)
plt.title('ETF')
plt.show()
```

```python
fig, ax=plt.subplots(figsize=(10,5))

ax.scatter(df['risk'][0],df['return'][0],color='red',marker='*',s=200,label='0050 ETF',alpha=0.5)
ax.scatter(df['risk'][1],df['return'][1],color='blue',marker='o',s=300,label='High Dividend ETF',alpha=0.5)
ax.scatter(df['risk'][2],df['return'][2],color='black',marker='v',s=400,label='G ETF',alpha=0.5)
ax.scatter(df['risk'][3],df['return'][3],color='pink',marker='8',s=500,label='Gold ETF',alpha=0.5)
ax.scatter(df['risk'][4],df['return'][4],color='yellow',marker='h',s=200,label='Oilx2 ETF',alpha=0.5)
ax.scatter(df['risk'][5],df['return'][5],color='green',marker='*',s=200,label='VIX ETF',alpha=0.5)
ax.scatter(df['risk'][6],df['return'][6],color='purple',marker='s',s=200,label='EMbond ETF',alpha=0.5)

plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.xlim(xmin=0,xmax=60)
plt.ylim(ymin=0)
plt.legend(loc='lower right')
plt.xlabel('risk',size=15)
plt.ylabel('return',size=15)
plt.title('ETF')
plt.show()
```
