## Part 1 靜態2D繪圖
在第一個部份，我們主要使用matplotlib來進行2D靜態圖的繪製。

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('seaborn')
mpl.__version__
```

### 繪製單維度的資料集
在這邊，單維度的資料集便是一組有兩組對應的序列所組成的資料集，像是股價的資料，我們有時間與股價這兩組序列，這應該是我們遇到最基本的資料集。
但我們要繪製時，我們可以用基本的plot()方法將折線圖繪出。
下面是一個實例:

```python
np.random.seed(1000) 
y = np.random.standard_normal(20)
x = np.arange(len(y))  
print(y)
print(x)
plt.plot(x, y)
```

```python
# 其實不一定要輸入x的值，預設為一系列整數
plt.plot(y)
```

```python
#我們也可以改變圖的一些設定，在plt下面有一系列函數可以更改繪畫風格的設定
plt.plot(y)
plt.grid(False)
plt.axis('equal')
```

plt.axis的一些指令
- equal --> 讓兩軸有相同的尺度
- tight --> 盡量讓所有的資料點畫在圖上
- off --> 不要顯示兩軸刻度

```python
# 另外兩個常用的繪畫風格指令為限制x軸與y軸的範圍
plt.plot(y)
plt.xlim(10,20)
plt.ylim(np.min(y)-1,np.max(y)+1)
```

```python
#為了繪圖的易讀性，我們會加入x軸與y軸的標籤，以及這張圖的標題
plt.figure(figsize=(10,6))
plt.plot(y)
plt.xlabel('index')
plt.ylabel('value')
plt.title('A plot')
```

```python
#我們也可以更改繪製的顏色
plt.figure(figsize=(10,6))
plt.plot(y,'black')
plt.xlabel('index')
plt.ylabel('value')
plt.title('A plot')
```

常用的一些顏色代號縮寫
- b --> 藍色
- g --> 綠色
- r --> 紅色
- k --> 黑色
- w --> 白色
- c --> 青色

```python
# plot函數預設是畫連續的直線，但是我們也可只畫資料點
plt.figure(figsize=(10,6))
plt.plot(y,'ro')
plt.xlabel('index')
plt.ylabel('value')
plt.title('A plot')
```

```python
#與畫連續線的結合使用
plt.figure(figsize=(10,6))
plt.plot(y)
plt.plot(y,'ro')
plt.xlabel('index')
plt.ylabel('value')
plt.title('A plot')
```

## Part2 多維度資料
假設，我們有幾隻股票的價格資料，我們想要在一張圖上表現它們的行為，我們要如何繪圖？

```python
y_1 = y
y_2 = y.cumsum()
print(y_1)
print(y_2)
```

```python
plt.figure(figsize=(10,6))
plt.plot(y_1,'black')
plt.plot(y_1,'ro')
plt.plot(y_2,'blue')
plt.plot(y_2,'ro')
plt.xlabel('index')
plt.ylabel('value')
plt.title('A plot')
```

```python
# 我們可以為兩條線加入說明
plt.figure(figsize=(10,6))
plt.plot(y_1,'black',label='1st')
plt.plot(y_1,'ro')
plt.plot(y_2,'blue',label='2nd')
plt.plot(y_2,'ro')
plt.xlabel('index')
plt.ylabel('value')
plt.legend()
plt.title('A plot')
```

```python
#假設我們現在兩個y值的尺度相差太大
plt.figure(figsize=(10,6))
plt.plot(y_1,'black',label='1st')
plt.plot(y_1,'ro')
plt.plot(y_2*100,'blue',label='2nd')
plt.plot(y_2*100,'ro')
plt.xlabel('index')
plt.ylabel('value')
plt.legend()
plt.title('A plot')
# 可以看到第一個值的變化無法準確的呈現
```

```python
fig, ax1 = plt.subplots()  
ax1.plot(y_1, 'b', lw=1.5, label='1st')
ax1.plot(y_1, 'ro')
plt.legend()
plt.xlabel('index')
plt.ylabel('value 1st')
plt.title('A Simple Plot')
ax2 = ax1.twinx()  
ax2.plot(y_2*100, 'g', lw=1.5, label='2nd')
ax2.plot(y_2*100, 'ro')
plt.legend()
plt.ylabel('value 2nd');
```

```python
#與上面不同，我們現在不用twinx來共享x軸
fig, (ax1,ax2) = plt.subplots(2,1)  
ax1.plot(y_1, 'b', lw=1.5, label='1st')
ax1.legend()
ax1.plot(y_1, 'ro')
ax1.set_ylabel('value 1st')
ax2.plot(y_2*100, 'g', lw=1.5, label='2nd')
ax2.plot(y_2*100, 'ro')
ax2.legend()
ax2.set_ylabel('value 2nd')
plt.title('A Simple Plot')
```

```python

fig, (ax1,ax2) = plt.subplots(1,2)  
ax1.plot(y_1, 'b', lw=1.5, label='1st')
ax1.legend()
ax1.plot(y_1, 'ro')
ax1.set_ylabel('value 1st')
ax2.plot(y_2*100, 'g', lw=1.5, label='2nd')
ax2.plot(y_2*100, 'ro')
ax2.legend()
ax2.set_ylabel('value 2nd')
plt.title('A Simple Plot')
```

### 其他種類圖形
- 散布圖
- 長條圖
- 箱形圖

```python
# 散布圖
y_11 = np.random.standard_normal(1000)
y_22 = np.random.standard_normal(1000)
plt.figure(figsize=(10,6))
plt.scatter(y_11,y_22,marker='o')
plt.xlabel('1st')
plt.ylabel('2nd')
plt.title('Scatter Plot')
```

```python
#長條圖
plt.figure(figsize=(10,6))
plt.hist([y_11,y_22],label=['1st','2nd'],bins=25)
plt.xlabel('value')
plt.ylabel('frequency')
plt.title('Histogram')
plt.legend()
```

```python
#箱圖
fig, ax = plt.subplots(figsize=(10, 6))
plt.boxplot([y_11,y_22])
plt.setp(ax,xticklabels=['1st', '2nd'])
plt.xlabel('data set')
plt.ylabel('value')
plt.title('Boxplot')
```
